```python
# Useful additional packages 
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from math import pi

from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit import BasicAer, execute
from qiskit.quantum_info import Pauli, state_fidelity, basis_state, process_fidelity 
```

```python
q = QuantumRegister(3, 'q')
```

```python
circ = QuantumCircuit(q)
circ.h(q[0])
circ.barrier()
circ.cu1(pi/2,q[0], q[1])
circ.barrier()
circ.cu1(pi/4,q[0], q[2])
circ.barrier()
circ.h(q[1])
circ.barrier()
circ.cu1(pi/2,q[1], q[2])
circ.barrier()
circ.h(q[2])

circ.draw()
```

```python
circ.draw(output='mpl')
```

```python
backend_sim = BasicAer.get_backend('statevector_simulator')

result = execute(circ, backend_sim).result()
state = result.get_statevector(circ)
print(state)
```

```python
q1 = QuantumRegister(4, 'q1')
```

```python
circ1 = QuantumCircuit(q1)

circ1.h(q1[0])
circ1.cu1(pi/2,q1[0], q1[1])
circ1.cu1(pi/4,q1[0], q1[2])
circ1.cu1(pi/8,q1[0], q1[3])

circ1.h(q1[1])
circ1.cu1(pi/2,q1[1], q1[2])
circ1.cu1(pi/4,q1[1], q1[3])

circ1.h(q1[2])
circ1.cu1(pi/2,q1[2], q1[3])

circ1.h(q1[3])

circ1.draw()
```

```python
circ1.draw(output='mpl')
```

```python
result1 = execute(circ1, backend_sim).result()
state1 = result1.get_statevector(circ1)
print(state1)
```

```python
backend1 = BasicAer.get_backend('unitary_simulator')
job1 = execute(circ1, backend1)
result2 = job1.result()
print(result2.get_unitary(circ1, decimals=3))
```
